```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Câu 1: Tạo DataFrame df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. Các danh sách
#này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv
df=pd.read_csv('sinh_vien_CN.csv')
b=pd.read_csv('sinh_vien_TO.csv')
c=pd.read_csv('sinh_vien_VL.csv')
```

```python
df=df.append(b)
df=df.append(c)
```

```python
#câu 4 tạo dataframe từ file kết quả 
df_ket_qua=pd.read_csv('ket_qua.csv')
df_ket_qua.head()
```

```python
#câu 5. Tạo DataFrame df_dtb bằng cách nhóm df_ket_qua theo masv và tính điểm trung bình cho mỗi sinh viên.
df_dtb = df_ket_qua.groupby('masv')['diem'].mean().to_frame().reset_index()
df_dtb
```

```python
# câu 6: Trộn df và df_dtb theo masv và có kết quả của df
df=pd.merge(df, df_dtb)
df
```

```python
# câu 7: In ra các sinh viên của khoa CN và có diem>8
df[(df['makh']=='CN') & (df['diem']>8)]
```

```python
#câu 8: In ra các sinh viên của khoa CN và có diem>8. Sắp xếp tăng theo ten và giảm theo diem
df[(df['makh']=='CN') & (df['diem']>8)].sort_values(by=['ten','diem'], ascending=[True,False])
```

```python
#câu 9: In ra các nam sinh viên không có học bổng. (câu này chưa làm được)
df[(df['hocbong']==0) & (df['gioitinh']==True)]
```

```python
#Câu 10: In ra tổng học bổng của mỗi khoa.
hoc_bong = df.groupby('makh')['hocbong'].sum().to_frame().reset_index()
hoc_bong
```

```python
#câu 13: Cho biết sinh viên có điểm lớn nhất
sap_xep =df.sort_values(by='diem', ascending=False) 
sap_xep.head(1)
```

```python
#câu 14: Cho biết 5 sinh viên có điểm lớn nhất mỗi khoa 
print('Top 5 sinh viên toán có điểm cao nhất')
df[df['makh']=='TO'].sort_values(by='diem',ascending=False).head()
```

```python
print('Top 5 sinh viên vật lý có điểm cao nhất')
df[df['makh']=='VL'].sort_values(by='diem',ascending=False).head()
```

```python
print('Top 5 sinh viên CN có điểm cao nhất')
df[df['makh']=='CN'].sort_values(by='diem',ascending=False).head()
```

```python
#Câu 15: Trong DataFrame df tạo thêm cột ketqua với kết quả là giỏi nếu điểm>=9, kết quả là khá nếu
#điểm<9 và điểm >=6.5, kết quả là trung bình nếu điểm <6.5 và điểm >=5, kết quả là yếu nếu
#điểm<5.
df['ketqua'] ='Yếu'
df.loc[df['diem'] >=9,'ketqua'] ='Giỏi'
df.loc[(df['diem'] >=6.5) & (df['diem'] <9),'ketqua'] ='Khá'
df.loc[(df['diem'] >=5) & (df['diem'] <6.5),'ketqua'] ='Trung Bình'
```

```python
#Trong mỗi khoa, đếm số sinh viên theo các loại kết quả
df.groupby('makh')['ketqua'].value_counts().to_frame()
```

```python
#câu 16 đếm số sv từng loại theo từng khoa 
df.groupby('makh')['ketqua'].value_counts().to_frame().rename(columns={'ketqua':'Số Lượng'})
```

```python
#câu 17. Vẽ biểu đồ boxplot với dữ liệu là DataFrame df.
sns.boxplot(data=df, y=df.diem)
plt.show()
```

```python
#Câu 17: Chép các outlier ra df_outliers: 
mo_ta=df.describe()
Q3=mo_ta['diem'][6]
Q1=mo_ta['diem'][4]
upper= Q3 + (Q3-Q1)*1.5
lower= Q1-(Q3-Q1)*1.5
df_outliers=df.loc[(df['diem'] < lower) | (df['diem'] > upper)]
df_outliers
```

```python
#câu 18: Vẽ biểu đồ boxplot theo mẫu sau và cho nhận xét
sns.boxplot(data=df, y='diem', x='makh')
#Nhận xét: 
# - Có thể thấy điểm trung bình của sinh viên Toán là cao nhất, kế đến là Vật Lý cuối cùng là Công Nghệ
# - Điểm cao nhất của cả 3 môn đều trên 8.0, môn CN có điểm cao nhất là lớn nhất trong 3 môn (8,75) 
# - Trong 3 môn không có sinh viên nào loại giỏi. 
# - Cả 3 môn đều có 50% số lượng sinh viên có khoảng điểm từ 7,75 đến 6,75. 
# - Môn vật lý có 2 sinh viên có điểm cực thấp là 5,5 và dưới 5,5. 
# - Môn CN có 1 bạn sinh viên có điểm cực cao (cao nhất trong tất cả sinh viên) là 8.75.
# - Tất cả sinh viên Toán đều có số điểm trong khoảng bình thường, không xuất hiện bất kỳ outliers nào trong môn Toán
```

```python
#Câu 19: Vẽ biểu đồ sau và cho biết sinh viên có điểm lớn nhất, nhỏ nhất theo mỗi khoa:
sns.swarmplot(data=df, x='makh',y= 'diem') 
```

```python
#Sinh viên có điểm thấp nhất khoa Vật Lý:
vl_min=df.loc[df['makh']=='VL']['diem'].min()
df.loc[(df['makh']=='VL') & (df['diem'] == vl_min)]
```

```python
#sinh viên có điểm lớn nhất khoa Vật Lý:
vl_max=df.loc[df['makh']=='VL']['diem'].max()
df.loc[(df['makh']=='VL') & (df['diem'] == vl_max)]
```

```python
#sinh viên có điểm thấp nhất khoa CN:
cn_min=df.loc[df['makh']=='CN']['diem'].min()
df.loc[(df['makh']=='CN') & (df['diem'] == cn_min)]
```

```python
#sinh viên có điểm cao nhất khoa CN:
cn_max=df.loc[df['makh']=='CN']['diem'].max()
df.loc[(df['makh']=='CN') & (df['diem'] == cn_max)]
```
